# Uso de Motores de Búsqueda Académicos

## Introducción

Los motores de búsqueda académicos son herramientas esenciales para la investigación científica. Permiten acceder a una amplia gama de publicaciones académicas, incluyendo artículos de revistas, tesis, libros y conferencias. En esta clase, nos centraremos en tres de los más importantes: Google Scholar, PubMed y Web of Science.


## 1. Google Scholar

[Google Scholar](https://scholar.google.com/) es un buscador de Google enfocado en el mundo académico.

### Características principales:
- Búsqueda de artículos de revistas científicas, tesis, libros, resúmenes y artículos de editoriales académicas.
- Muestra quién ha citado cada documento.
- Permite crear un perfil de investigador.

### Técnicas de búsqueda avanzada:
- Uso de comillas para búsqueda exacta: "cambio climático"
- Operadores booleanos: "cambio climático" AND "efectos económicos"
- Búsqueda por autor: author:"J Smith"
- Búsqueda por fecha: after:2020

### Ejercicio práctico:
1. Visita [Google Scholar](https://scholar.google.com/)
2. Realiza una búsqueda sobre un tema de tu interés utilizando las técnicas avanzadas.
3. Configura una alerta para recibir nuevas publicaciones sobre este tema.


### a) Búsqueda por frase exacta:
- Uso: Pon la frase entre comillas
- Ejemplo: "cambio climático en América Latina"
- Esto buscará exactamente esa frase, en ese orden.

### b) Exclusión de palabras:
- Uso: Utiliza el signo menos (-) antes de la palabra que quieres excluir
- Ejemplo: cambio climático -Europa
- Esto buscará artículos sobre cambio climático, pero excluirá aquellos que mencionen Europa.

### c) Búsqueda por autor:
- Uso: Utiliza "author:" antes del nombre del autor
- Ejemplo: author:"J González" cambio climático
- Esto buscará artículos sobre cambio climático escritos por autores con el nombre J González.

### d) Búsqueda por año o rango de años:
- Uso: Utiliza los comandos after: y before:
- Ejemplo: "cambio climático" after:2020
- Esto buscará artículos sobre cambio climático publicados después de 2020.

### e) Búsqueda en el título:
- Uso: Utiliza "allintitle:"
- Ejemplo: allintitle: "inteligencia artificial" educación
- Esto buscará artículos que tengan tanto "inteligencia artificial" como "educación" en el título.

### Ejemplo de búsqueda combinada:
"inteligencia artificial" educación author:"A García" after:2018 -"machine learning"

Esta búsqueda encontrará artículos que:
- Contengan la frase exacta "inteligencia artificial"
- Mencionen educación
- Estén escritos por un autor llamado A García
- Hayan sido publicados después de 2018
- No contengan la frase "machine learning"


## 2. Web of Science: Búsquedas Avanzadas


[Web of Science](https://www.webofscience.com/) es una plataforma que incluye múltiples bases de datos de citas.

### Características principales:
- Cobertura multidisciplinaria
- Proporciona métricas de impacto de las publicaciones
- Permite análisis de citas

### Búsqueda avanzada:
1. Accede a la [búsqueda avanzada de Web of Science](https://www.webofscience.com/wos/woscc/advanced-search)
2. Utiliza los campos de búsqueda para combinar diferentes criterios

### Ejercicio práctico:
1. Accede a Web of Science (requiere suscripción, generalmente disponible a través de instituciones académicas)
2. Realiza una búsqueda sobre tu tema de interés
3. Analiza las citas de uno de los artículos encontrados

### a) Búsqueda por tema y título:
- En el campo "Add row", selecciona "Topic" y "Title" en dos filas separadas
- Ejemplo:
  Topic: "artificial intelligence"
  Title: education
- Esto buscará artículos que tengan "artificial intelligence" en cualquier parte y "education" en el título.

### b) Búsqueda por autor y año:
- Usa los campos "Author" y "Year Published"
- Ejemplo:
  Author: García, A
  Year Published: 2018-2023
- Esto buscará artículos de A. García publicados entre 2018 y 2023.

### c) Búsqueda por organización y país:
- Usa los campos "Organization-Enhanced" y "Country/Region"
- Ejemplo:
  Organization-Enhanced: Universidad de Buenos Aires
  Country/Region: Argentina
- Esto buscará artículos afiliados a la Universidad de Buenos Aires en Argentina.

### d) Búsqueda por tipo de documento y idioma:
- Usa los campos "Document Type" y "Language"
- Ejemplo:
  Document Type: Article
  Language: Spanish
- Esto buscará artículos de revista publicados en español.

# Búsqueda Avanzada en Web of Science: Efectos de El Niño en las costas de Chile (2019-2024)

Para realizar esta búsqueda en Web of Science, sigue estos pasos:

1. Accede a [Web of Science](https://www.webofscience.com/)
2. Haz clic en "Búsqueda avanzada" o "Advanced search"
3. Utiliza la siguiente estructura de búsqueda:

```
TS=("El Niño" OR "ENSO") AND TS=(coast* OR shore* OR litoral) AND TS=(Chile)
AND PY=(2019-2024)
```

Desglose de la búsqueda:

- `TS=`: Busca en el campo "Topic", que incluye título, resumen y palabras clave.
- `"El Niño" OR "ENSO"`: Busca artículos que mencionen El Niño o su acrónimo científico ENSO (El Niño Southern Oscillation).
- `coast* OR shore* OR litoral`: Busca términos relacionados con la costa. El asterisco (*) permite incluir variaciones como "coastal", "coastline", etc.
- `Chile`: Limita la búsqueda a estudios relacionados con Chile.
- `PY=(2019-2024)`: Limita la búsqueda a publicaciones entre 2019 y 2024.

Opciones adicionales para refinar la búsqueda:

1. Tipo de documento:
   Añade: `AND DT=(Article OR Review)`
   Esto limitará los resultados a artículos y revisiones.

2. Idioma:
   Añade: `AND LA=(English OR Spanish)`
   Esto limitará los resultados a publicaciones en inglés o español.

3. Categorías de Web of Science:
   Añade: `AND WC=(Oceanography OR Meteorology & Atmospheric Sciences OR Environmental Sciences)`
   Esto enfocará la búsqueda en categorías relevantes.

Búsqueda completa con todas las opciones:

```
TS=("El Niño" OR "ENSO") AND TS=(coast* OR shore* OR litoral) AND TS=(Chile)
AND PY=(2019-2024) AND DT=(Article OR Review) AND LA=(English OR Spanish)
AND WC=(Oceanography OR Meteorology & Atmospheric Sciences OR Environmental Sciences)
```

Después de ejecutar la búsqueda, puedes usar los filtros del lado izquierdo para refinar aún más los resultados por institución, autor, revista, etc.





## Comparación y mejores prácticas

- Google Scholar: Amplia cobertura, incluye citas, gratuito. Mejor para búsquedas generales y multidisciplinarias.
- PubMed: Especializado en ciencias de la vida y biomedicina. Mejor para búsquedas en estas áreas.
- Web of Science: Multidisciplinario, con análisis de citas detallado. Mejor para análisis de impacto y búsquedas muy específicas.

### Estrategia recomendada:
1. Comienza con Google Scholar para una visión general
2. Utiliza PubMed o Web of Science para profundizar en áreas específicas
3. Compara los resultados de las diferentes plataformas
