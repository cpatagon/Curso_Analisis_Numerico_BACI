# Guía para la Elaboración de Póster Científico en Ciencias Ambientales

## 1. Fundamentos del Póster Científico Ambiental

### 1.1 Propósito
- Comunicar investigaciones ambientales de manera visual y efectiva
- Presentar resultados de monitoreos ambientales
- Mostrar relaciones entre variables ecosistémicas
- Comunicar problemáticas ambientales y sus soluciones

### 1.2 Tipos de Investigaciones Aplicables
- Estudios de impacto ambiental
- Monitoreo de ecosistemas
- Análisis de contaminación
- Estudios de biodiversidad
- Cambio climático
- Gestión de recursos naturales
- Restauración ecológica

## 2. Estructura del Póster

### 2.0 Estrucura general


### 2.1 Título y Autores
#### Título Efectivo
- Longitud óptima: 10-15 palabras
- Incluir: qué, dónde, cuándo
- Destacar el aspecto innovador

Ejemplos:

- ✓ "Evaluación temporal de la calidad del aire urbano mediante análisis multivariado: Ciudad de Santiago 2020-2024"
- ✓ "Patrones de diversidad vegetal en gradientes altitudinales: Impactos del cambio climático"
- ✓ "Bioindicadores de contaminación hídrica en cuencas agrícolas: Un enfoque integrado"

#### Información de Autores
- Nombres y afiliaciones
- Correo electrónico de contacto
- Institución y departamento
- Códigos QR para información adicional

### 2.2 Introducción

¿qué problema?

- Contexto ambiental
- Problemática específica
- Relevancia para la gestión/conservación
- Objetivos claros (¿para qué?)
- Hipótesis (si aplica)

### 2.3 Metodología

¿Cómo se estudio?

#### Diseño del Estudio
- Área de estudio
- Período de muestreo
- Variables medidas
- Equipos utilizados
- Protocolos estandarizados

#### Análisis de Datos
- Métodos estadísticos
- Software utilizado
- Modelos aplicados
- Criterios de calidad

### 2.4 Resultados

¿Cuales fueron sus hallazgos?

#### Tipos de Visualización
1. Datos Espaciales:
   - Mapas de distribución
   - Gradientes ambientales
   - Zonas de impacto

2. Datos Temporales:
   - Series de tiempo
   - Patrones estacionales
   - Tendencias a largo plazo

3. Relaciones entre Variables:
   - Gráficos de dispersión
   - Análisis multivariados
   - Redes de interacción

4. Comparaciones:
   - Gráficos de barras
   - Diagramas de caja
   - Gráficos de violín

### 2.5 Discución y/o Conclusiones

¿Qué significado tienen?

- Hallazgos principales
- Implicaciones prácticas
- Recomendaciones
- Próximos pasos

## 3. Diseño Visual

### 3.1 Especificaciones Técnicas
- Tamaño: 90x120 cm (vertical preferentemente)
- Resolución: 300 dpi mínimo
- Formato final: PDF de alta calidad

### 3.2 Organización y Flujo
```
+--------------------------------+
|             TÍTULO             |
|             Autores            |
+---------------+----------------+
|  Introducción | • Metodología  |
|  • Contexto   | • Diseño       |
|  • Objetivos  | • Análisis     |
+---------------+----------------+
|         RESULTADOS            |
| • Gráficos    • Tablas        |
| • Mapas       • Estadísticas   |
+---------------+----------------+
| Conclusiones  | Referencias    |
+---------------+----------------+
|        Información de          |
|          Contacto             |
+--------------------------------+
```

### 3.3 Elementos Visuales
#### Tipografía
- Títulos: 72-84 pt
- Subtítulos: 48-60 pt
- Texto principal: 32-36 pt
- Pies de figura: 24-28 pt
- Notas: 18-20 pt

#### Paleta de Colores
1. Esquema Profesional:
   - Verde oscuro (#2E5A35)
   - Verde medio (#568F59)
   - Verde claro (#88B78D)
   - Gris (#4A4A4A)

2. Esquema de Datos:
   - Usar paletas colorblind-friendly
   - Mantener consistencia en gráficos
   - Evitar rojo-verde juntos

## 4. Visualización de Datos Ambientales

### 4.1 Mejores Prácticas
- Escalas apropiadas
- Unidades claras
- Barras de error
- Leyendas completas
- Códigos de color consistentes

### 4.2 Tipos de Gráficos por Variable
1. Variables Continuas:
   - Series temporales
   - Gráficos de dispersión
   - Superficies de respuesta

2. Variables Categóricas:
   - Gráficos de barras
   - Diagramas de torta
   - Gráficos de mosaico

3. Datos Espaciales:
   - Mapas temáticos
   - Mapas de calor
   - Interpolaciones

### 5.2 Visualización

### 5.3 Técnico

## 6. Rúbrica de Evaluación

| Criterio | Peso | Aspectos Evaluados |
|----------|------|-------------------|
| Contenido | 40% | - Rigor científico<br>- Claridad metodológica<br>- Validez de resultados |
| Diseño | 25% | - Organización<br>- Legibilidad<br>- Estética |
| Visualización | 25% | - Calidad de gráficos<br>- Claridad de datos<br>- Efectividad visual |
| Técnico | 10% | - Formato<br>- Referencias<br>- Aspectos formales |

## 7. Consejos Finales
1. **Preparación**:
   - Revisar múltiples veces
   - Imprimir prueba pequeña
   - Verificar colores

2. **Presentación**:
   - Preparar resumen de 2 minutos
   - Anticipar preguntas comunes
   - Tener tarjetas de contacto

3. **Documentación**:
   - Guardar archivos editables
   - Mantener respaldo de datos
   - Documentar procedimientos
