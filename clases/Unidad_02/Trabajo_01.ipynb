# Guía de trabajo: Barrido bibliográfico

## Objetivo
Realizar un barrido bibliográfico sobre una temática de interés en el campo de la biología marina, que permita construir un problema a abordar, definir un objetivo y obtener datos relevantes para su análisis.

## Pasos a seguir

### 1. Selección de la temática
- Escoge un tema específico dentro del análisis numérico que te interese profundizar.


### 2. Búsqueda inicial de fuentes
- Utiliza bases de datos académicas como Web of Science, Scopus, o Google Scholar.
- Busca artículos científicos, libros de texto y tesis relacionados con tu tema.
- Identifica palabras clave relevantes para refinar tu búsqueda.
- acote su busqueda de acuerdo a año autor, ubicación geografica etc.

### 3. Lectura y análisis preliminar
- Revisa los resúmenes y conclusiones de los documentos encontrados.
- Selecciona al menos 10 fuentes relevantes y actuales (preferiblemente de los últimos 5 años).
- Identifica los autores y revistas más prominentes en el área.

### 4. Construcción del problema a abordar
- Basándote en la literatura revisada, identifica:
  - Desafíos actuales en el campo
  - Limitaciones de los métodos existentes
  - Áreas poco exploradas o con potencial de mejora
- Formula una pregunta de investigación clara y concisa.

### 5. Definición del objetivo
- Establece un objetivo general que responda a la pregunta de investigación.
- Define 2-3 objetivos específicos que contribuyan al logro del objetivo general.

### 6. Recopilación de datos relevantes
- Extrae de tus fuentes:
  - Métodos utilizados
  - Conjuntos de datos empleados
  - Métricas de evaluación comunes
  - Resultados y conclusiones principales
- Organiza esta información en una tabla comparativa.

### 7. Análisis crítico
- Compara y contrasta los diferentes enfoques encontrados.
- Identifica fortalezas y debilidades de cada método o enfoque.
- Reconoce patrones o tendencias en la investigación reciente.

### 8. Síntesis y conclusiones
- Resume los hallazgos principales de tu barrido bibliográfico.
- Justifica la relevancia y originalidad del problema que has construido.
- Explica cómo tu objetivo contribuirá al avance del conocimiento en el área.

### 9. Elaboración del informe final
- Introduce el tema y su importancia en el análisis numérico.
- Describe tu metodología de búsqueda y selección de fuentes.
- Presenta el problema construido y tu objetivo de manera clara.
- Expón los datos relevantes obtenidos y tu análisis crítico.
- Concluye con la síntesis de tus hallazgos y propuestas para futuras investigaciones.

## Entregables
1. Informe escrito (máximo 3000 palabras) que incluya todos los elementos mencionados.
2. Bibliografía en formato APA o IEEE de las fuentes utilizadas.
3. Tabla comparativa de los datos relevantes extraídos.
4. Presentación breve (5-7 diapositivas) resumiendo los puntos clave de tu trabajo.

## Evaluación
Tu trabajo será evaluado según los siguientes criterios:
- Relevancia y actualidad de las fuentes seleccionadas
- Claridad en la construcción del problema y definición del objetivo
- Calidad del análisis crítico y síntesis de la información
- Organización y presentación del informe final
- Originalidad y potencial de la propuesta de investigación
