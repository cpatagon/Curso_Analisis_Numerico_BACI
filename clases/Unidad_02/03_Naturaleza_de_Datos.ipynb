# Entendiendo la Naturaleza de los Datos

El objeto de estudio del análisis de datos son, básicamente, los datos. 

Los datos son entonces los actores clave en todos los procesos del análisis de datos. 

Constituyen la materia prima que debe ser procesada, y gracias a su procesamiento y análisis es posible extraer una variedad de información para aumentar el nivel de conocimiento del sistema en estudio, es decir, aquel del que provienen los datos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generar datos oceanográficos de ejemplo
data = {
    'fecha': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'temperatura': [20 + i*0.1 + np.random.randn() for i in range(100)],
    'salinidad': [35 + i*0.01 + np.random.randn()*0.1 for i in range(100)],
    'profundidad': [i*10 for i in range(100)],
}

df = pd.DataFrame(data)

print(df.head(10))
```

## Cuando la Información se Convierte en Conocimiento

Se puede hablar de conocimiento cuando la información se convierte en un conjunto de reglas que te ayudan a entender mejor ciertos mecanismos y, por lo tanto, a hacer predicciones sobre la evolución de algunos eventos.

Veamos cómo podemos convertir nuestros datos en información útil:

### Visualizar la relación entre temperatura y salinidad

```python

plt.figure(figsize=(10, 6))
plt.scatter(df['temperatura'], df['salinidad'])
plt.xlabel('Temperatura (°C)')
plt.ylabel('Salinidad (PSU)')
plt.title('Relación entre Temperatura y Salinidad')
plt.show()
```

```python
from scipy import stats

# Calcular la correlación entre temperatura y salinidad
correlation = df['temperatura'].corr(df['salinidad'])
print(f"Correlación entre temperatura y salinidad: {correlation}")

# Realizar una regresión lineal
slope, intercept, r_value, p_value, std_err = stats.linregress(df['temperatura'], df['salinidad'])

print(f"Pendiente: {slope}")
print(f"Intercepto: {intercept}")
print(f"Valor R: {r_value}")
print(f"Valor P: {p_value}")
```





## Tipos de Datos

Los datos pueden dividirse en dos categorías distintas:

- **Categóricos**
  - **Nominales**: Una variable nominal no tiene un orden intrínseco en su categoría.
  - **Ordinales**: Una variable ordinal tiene un orden predeterminado.
  
- **Numéricos**
  - **Discretos**: Valores que pueden contarse y que son distintos y separados entre sí.
  - **Continuos**: Valores producidos por mediciones u observaciones que pueden asumir cualquier valor dentro de un rango definido.

Los datos categóricos son valores u observaciones que pueden dividirse en grupos o categorías. Hay dos tipos de valores categóricos: nominales y ordinales.

Los datos numéricos son valores u observaciones que provienen de mediciones. Hay dos tipos de valores numéricos diferentes: números discretos y continuos. Los valores discretos son valores que pueden contarse y que son distintos y separados entre sí. Los valores continuos, por otro lado, son valores producidos por mediciones u observaciones que asumen cualquier valor dentro de un rango definido.


# Tipos de Datos en Oceanografía

En oceanografía, nos encontramos con diversos tipos de datos. Entender estos tipos es crucial para el análisis y la interpretación correcta de la información oceánica. Veamos cada tipo con ejemplos específicos del campo:

## Datos Categóricos

### Nominales
Una variable nominal no tiene un orden intrínseco en su categoría.

Ejemplos en oceanografía:
1. Tipos de masas de agua
2. Especies marinas
3. Zonas oceánicas


```python

import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con datos nominales
data_nominal = {
    'masa_agua': ['NADW', 'AAIW', 'NADW', 'AABW', 'AAIW', 'NADW'],
    'especie_marina': ['Atún', 'Sardina', 'Tiburón', 'Delfín', 'Atún', 'Sardina'],
    'zona_oceanica': ['Epipelágica', 'Mesopelágica', 'Batipelágica', 'Epipelágica', 'Mesopelágica', 'Batipelágica']
}

df_nominal = pd.DataFrame(data_nominal)

# Analizar datos nominales
print(df_nominal['masa_agua'].value_counts())

# Visualizar datos nominales
plt.figure(figsize=(10, 6))
df_nominal['zona_oceanica'].value_counts().plot(kind='bar')
plt.title('Distribución de Zonas Oceánicas')
plt.ylabel('Frecuencia')
plt.show()
```

### Ordinales
Una variable ordinal tiene un orden predeterminado.

Ejemplos en oceanografía:
1. Escala de Beaufort (fuerza del viento)
2. Niveles de contaminación marina
3. Categorías de tamaño de sedimentos

```python
import numpy as np

# Crear un DataFrame con datos ordinales
data_ordinal = {
    'escala_beaufort': [0, 3, 6, 2, 8, 11],
    'nivel_contaminacion': ['Bajo', 'Medio', 'Alto', 'Bajo', 'Muy Alto', 'Medio'],
    'tamano_sedimento': ['Arcilla', 'Limo', 'Arena', 'Grava', 'Canto', 'Arena']
}

df_ordinal = pd.DataFrame(data_ordinal)

# Definir el orden para las variables ordinales
nivel_contaminacion_orden = ['Bajo', 'Medio', 'Alto', 'Muy Alto']
tamano_sedimento_orden = ['Arcilla', 'Limo', 'Arena', 'Grava', 'Canto']

# Convertir a tipo categórico con orden específico
df_ordinal['nivel_contaminacion'] = pd.Categorical(df_ordinal['nivel_contaminacion'], 
                                                   categories=nivel_contaminacion_orden, 
                                                   ordered=True)

# Analizar datos ordinales
print(df_ordinal['escala_beaufort'].value_counts().sort_index())

# Visualizar datos ordinales
plt.figure(figsize=(10, 6))
df_ordinal['nivel_contaminacion'].value_counts().sort_index().plot(kind='bar')
plt.title('Niveles de Contaminación Marina')
plt.ylabel('Frecuencia')
plt.show()
```


## Datos Numéricos

### Discretos
Valores que pueden contarse y que son distintos y separados entre sí.

Ejemplos en oceanografía:
1. Número de especies en un arrecife
2. Cantidad de boyas oceanográficas en una región
3. Número de capas en la columna de agua

```python
# Crear un DataFrame con datos discretos
data_discreto = {
    'num_especies_arrecife': [10, 15, 8, 22, 13, 18],
    'num_boyas': [5, 3, 7, 4, 6, 5],
    'num_capas_agua': [3, 2, 4, 3, 5, 2]
}

df_discreto = pd.DataFrame(data_discreto)

# Analizar datos discretos
print(df_discreto['num_especies_arrecife'].describe())

# Visualizar datos discretos
plt.figure(figsize=(10, 6))
plt.hist(df_discreto['num_especies_arrecife'], bins=range(min(df_discreto['num_especies_arrecife']), 
                                                         max(df_discreto['num_especies_arrecife']) + 2, 1))
plt.title('Distribución del Número de Especies en Arrecifes')
plt.xlabel('Número de Especies')
plt.ylabel('Frecuencia')
plt.show()
```

### Continuos
Valores producidos por mediciones u observaciones que pueden asumir cualquier valor dentro de un rango definido.

Ejemplos en oceanografía:
1. Temperatura del agua
2. Salinidad
3. Profundidad
4. Concentración de oxígeno disuelto

```python
# Crear un DataFrame con datos continuos
np.random.seed(42)
data_continuo = {
    'temperatura_agua': np.random.normal(20, 2, 100),  # Media 20°C, desviación estándar 2°C
    'salinidad': np.random.normal(35, 0.5, 100),  # Media 35 PSU, desviación estándar 0.5 PSU
    'profundidad': np.random.uniform(0, 1000, 100),  # Profundidades entre 0 y 1000 metros
    'oxigeno_disuelto': np.random.normal(6, 1, 100)  # Media 6 mg/L, desviación estándar 1 mg/L
}

df_continuo = pd.DataFrame(data_continuo)

# Analizar datos continuos
print(df_continuo.describe())

# Visualizar datos continuos
fig, axs = plt.subplots(2, 2, figsize=(15, 15))
axs[0, 0].hist(df_continuo['temperatura_agua'], bins=20)
axs[0, 0].set_title('Distribución de Temperatura del Agua')
axs[0, 1].hist(df_continuo['salinidad'], bins=20)
axs[0, 1].set_title('Distribución de Salinidad')
axs[1, 0].hist(df_continuo['profundidad'], bins=20)
axs[1, 0].set_title('Distribución de Profundidad')
axs[1, 1].hist(df_continuo['oxigeno_disuelto'], bins=20)
axs[1, 1].set_title('Distribución de Oxígeno Disuelto')
plt.tight_layout()
plt.show()

# Relación entre variables continuas
plt.figure(figsize=(10, 6))
plt.scatter(df_continuo['temperatura_agua'], df_continuo['oxigeno_disuelto'])
plt.title('Relación entre Temperatura y Oxígeno Disuelto')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Oxígeno Disuelto (mg/L)')
plt.show()
```

Estos ejemplos ilustran cómo diferentes tipos de datos se presentan en la oceanografía y cómo podemos analizarlos y visualizarlos usando Python. Entender estos tipos de datos es crucial para:

1. Elegir los métodos estadísticos apropiados
2. Seleccionar las visualizaciones más efectivas
3. Interpretar correctamente los resultados del análisis
4. Diseñar experimentos y estudios oceanográficos
