# Las Publicaciones Científicas como Fuente de Datos

Las publicaciones científicas son fundamentales para la difusión del conocimiento y constituyen una valiosa **fuente de datos** en distintos campos. 

A través de revistas especializadas, artículos y estudios publicados, los investigadores pueden acceder a datos cruciales sobre ecosistemas acuáticos, especies marinas, procesos oceanográficos entre otros.

## Acceso a Fuentes de Datos a través de Publicaciones en Biología Marina

Muchas publicaciones en biología marina incluyen conjuntos de datos reutilizables, que pueden estar en formato tabular o accesibles mediante bases de datos enlazadas en los artículos. Plataformas como **Ocean Biodiversity Information System (OBIS)** o **World Register of Marine Species (WoRMS)** permiten acceder a publicaciones y datasets asociados.

### Ejemplos de tipos de datos disponibles en publicaciones de biología marina:

- Censos de poblaciones marinas
- Datos de telemetría de especies migratorias
- Series temporales de parámetros oceanográficos
- Registros de biodiversidad marina
- Datos sobre impacto del cambio climático en ecosistemas marinos

## Beneficios de Utilizar Datos de Publicaciones en Biología Marina:

1. **Validez y Relevancia:** Los datos en publicaciones marinas han pasado por revisión por pares, asegurando su calidad y pertinencia en el campo.

2. **Acceso Abierto:** Revistas como *PLOS ONE* ofrecen acceso libre a datos marinos, permitiendo su uso sin restricciones.

3. **Facilidad de Referencia:** Los artículos proporcionan metodología y contexto sobre la obtención de datos marinos, facilitando su interpretación y aplicación.

4. **Cobertura Geográfica:** Acceso a datos de ecosistemas marinos de diferentes regiones del mundo.

5. **Análisis Temporal:** Posibilidad de estudiar cambios a largo plazo en poblaciones y ecosistemas marinos.


# Cómo Buscar Datasets en Papers Científicos de Biología Marina

## 1. Identifica las Secciones Clave

Los datasets suelen mencionarse en secciones específicas de los papers:

- **Métodos / Materiales y Métodos**: Descripción de cómo se recolectaron los datos.

![](./img/datos_publicacion.png)

- **Resultados**: Pueden incluir tablas o figuras con datos.

![](./img/tablas_resultados.png)


- **Información Suplementaria**: Muchos datasets extensos se incluyen aquí.

![](./img/info_suplementaria.png)

- **Declaración de Disponibilidad de Datos**: Sección obligatoria en muchas revistas modernas.

## 2. Busca Palabras Clave

Palabras y frases que suelen indicar la presencia de datasets:

- "dataset"
- "data available at"
- "supplementary data"
- "raw data"
- "data repository"

## 4. Utiliza Herramientas de Búsqueda Especializadas

Algunas bases de datos facilitan la búsqueda de papers con datasets:

- [Elsevier's DataSearch](https://datasearch.elsevier.com/)
- [Google Dataset Search](https://datasetsearch.research.google.com/)

## 5. Contacta a los Autores

Si el dataset no está disponible públicamente:

1. Busca la información de contacto en el paper.
2. Envía un email cortés solicitando acceso a los datos.

## Ejemplo Práctico: Búsqueda de Datos sobre Arrecifes de Coral

1. Busca en Google Scholar: "coral reef biodiversity dataset"
2. Abre un artículo relevante, por ejemplo: "Global patterns of coral diversity and biomass in tropical reefs"
3. Ve a la sección "Methods" y busca menciones de datos
4. Revisa la "Data Availability Statement" al final del paper
5. Sigue los enlaces proporcionados



## Ejemplos de Uso de Datos de Publicaciones en Biología Marina

### 1. Estudio de Migraciones

Un investigador podría utilizar datos de telemetría publicados sobre ballenas jorobadas para analizar patrones de migración en relación con cambios en la temperatura del océano.

#### Análisis de Migración de Ballenas Jorobadas

##### Datos utilizados:
- Fuente: Smith et al. (2023), Marine Ecology Progress Series
- Tipo: Datos de telemetría satelital
- Periodo: 2018-2022
- Muestra: 50 ballenas jorobadas

##### Metodología:
1. Extracción de coordenadas geográficas y fechas
2. Cálculo de distancias recorridas
3. Correlación con datos de temperatura superficial del mar

### 2. Biodiversidad de Arrecifes

Un equipo de investigación podría combinar datos de censos de peces de varias publicaciones para crear un modelo de biodiversidad de arrecifes coralinos.

#### Modelo de Biodiversidad de Arrecifes Coralinos

##### Fuentes de Datos:
1. Johnson et al. (2021), Coral Reefs - Censo del Caribe
2. Lee et al. (2022), Journal of Marine Biology - Datos del Pacífico Sur
3. García et al. (2023), Biodiversity and Conservation - Arrecifes del Mar Rojo

##### Pasos del Análisis:
1. Estandarización de métodos de conteo entre estudios
2. Análisis de riqueza de especies por región
3. Modelado de factores que influyen en la biodiversidad

### 3. Impacto del Cambio Climático

Investigadores podrían utilizar series temporales publicadas sobre blanqueamiento de corales para predecir futuros impactos del calentamiento global.

#### Predicción de Blanqueamiento de Corales

##### Datos Históricos:
- Fuente: Global Coral Reef Monitoring Network (publicaciones anuales)
- Periodo: 1998-2023
- Cobertura: 100 arrecifes en todo el mundo

##### Análisis:
1. Tendencias de blanqueamiento por región
2. Correlación con anomalías de temperatura superficial del mar
3. Modelado predictivo utilizando escenarios climáticos del IPCC

Las publicaciones científicas en biología marina no solo difunden conocimiento, sino que también proporcionan **fuentes de datos esenciales** para análisis exhaustivos y bien fundamentados sobre la vida y los procesos en los océanos.
