# Mendeley: Software de Gestión Bibliográfica

## Introducción

Mendeley es una herramienta gratuita de gestión bibliográfica ampliamente utilizada en el ámbito académico. Este software ofrece numerosas funcionalidades que simplifican el proceso de investigación y organización de referencias.

- [sitio de Mendeley](https://www.mendeley.com/search/)

![](./img/mendeley.png)


## Características principales

### 1. Fácil instalación y uso
- Se instala rápidamente en tu computadora
- Incluye un ícono en la barra de navegación de Google para fácil acceso

### 2. Gestión de referencias
- Añade automáticamente artículos a tu biblioteca con un clic
- Para publicaciones indexadas, registra automáticamente:
  - Autores
  - Título
  - Año de publicación
  - Lugar de publicación
  - Cantidad de páginas
  - Abstract
- Permite añadir notas personales a los artículos

### 3. Organización
- Crea carpetas para ordenar tus referencias
- Similar a "Mi biblioteca" de Google Académico, pero con funcionalidades adicionales

### 4. Análisis de relevancia
Proporciona información valiosa sobre los artículos:
- Número de citaciones por año
- Perfil de los lectores (investigadores, estudiantes, profesores)
- Disciplina de los lectores
- Estado de indexación del artículo

### 5. Recomendaciones
- Sugiere artículos relacionados basándose en tu biblioteca
- Utiliza algoritmos para buscar temas en común, palabras clave, autores y citaciones cruzadas

### 6. Alertas personalizadas
- Envía correos electrónicos con nuevos artículos basados en tu biblioteca o criterios específicos
- Similar a las alertas de Google Académico

## Consejos de uso

1. **Lectura gradual**: Busca 2-3 artículos, léelos y analízalos antes de continuar la búsqueda. Esto ayuda a refinar las búsquedas posteriores.

2. **Evaluación crítica**: No todo lo útil estará en revistas prestigiosas o muy citado. Considera también:
   - Trabajos de instituciones locales (educativas, productivas)
   - Informes técnicos
   - Artículos no académicos relevantes para tu tema

3. **Balanceo de fuentes**: Encuentra un equilibrio entre fuentes altamente reconocidas y otras menos convencionales pero relevantes.

4. **Justificación**: Si incluyes fuentes menos reconocidas, prepárate para justificar su uso en tu investigación.

## Conclusión

Mendeley es una herramienta poderosa que puede mejorar significativamente tu proceso de investigación y gestión bibliográfica. Sin embargo, recuerda que es un complemento a tu criterio como investigador, no un sustituto. La evaluación crítica y el balance en la selección de fuentes siguen siendo fundamentales en el proceso de investigación.
