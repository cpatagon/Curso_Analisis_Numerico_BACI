# Búsqueda Avanzada en Web of Science: Efectos de El Niño en las costas de Chile (2019-2024)

Para realizar esta búsqueda en Web of Science, sigue estos pasos:

## 1. Accede a [Web of Science](https://www.webofscience.com/) 

e ingresa usuario y password


![imagen inicial](./img/webofscience.png)

## 2. Haz clic en "Búsqueda avanzada" o "Advanced search"

![imagen busqueda avanzada](./img/busqueda.png)

## 3. Utiliza la siguiente estructura de búsqueda:

```
TS=("El Niño" OR "ENSO") AND TS=(coast* OR shore* OR litoral) AND TS=(Chile)
AND PY=(2019-2024)
```

![busqueda avanzada](./img/busquedaavanzada.png)

Desglose de la búsqueda:

- `TS=`: Busca en el campo "Topic", que incluye título, resumen y palabras clave.
- `"El Niño" OR "ENSO"`: Busca artículos que mencionen El Niño o su acrónimo científico ENSO (El Niño Southern Oscillation).
- `coast* OR shore* OR litoral`: Busca términos relacionados con la costa. El asterisco (*) permite incluir variaciones como "coastal", "coastline", etc.
- `Chile`: Limita la búsqueda a estudios relacionados con Chile.
- `PY=(2019-2024)`: Limita la búsqueda a publicaciones entre 2019 y 2024.

Opciones adicionales para refinar la búsqueda:

1. Tipo de documento:
   Añade: `AND DT=(Article OR Review)`
   Esto limitará los resultados a artículos y revisiones.

2. Idioma:
   Añade: `AND LA=(English OR Spanish)`
   Esto limitará los resultados a publicaciones en inglés o español.

3. Categorías de Web of Science:
   Añade: `AND WC=(Oceanography OR Meteorology & Atmospheric Sciences OR Environmental Sciences)`
   Esto enfocará la búsqueda en categorías relevantes.

Búsqueda completa con todas las opciones:

```
TS=("El Niño" OR "ENSO") AND TS=(coast* OR shore* OR litoral) AND TS=(Chile)
AND PY=(2019-2024) AND DT=(Article OR Review) AND LA=(English OR Spanish)
AND WC=(Oceanography OR Meteorology & Atmospheric Sciences OR Environmental Sciences)
```

Después de ejecutar la búsqueda, puedes usar los filtros del lado izquierdo para refinar aún más los resultados por institución, autor, revista, etc.
